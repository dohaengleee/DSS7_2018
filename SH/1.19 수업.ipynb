```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
A = X[:]
b = y[:]
```

```python
x, resid, rank, s = np.linalg.lstsq(A, b)
x
```

```python
Apinv = np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)
Apinv
```

```python
i = np.dot(Apinv, b)
i
```

```python
np.dot(A, i)
```
