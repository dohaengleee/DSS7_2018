```python
i = np.array([[1, 0.5],[0.3,1]])
iInv = np.linalg.inv(i)
```

```python
ex1 = np.array([1, 0])
ex2 = np.array([0, 1])
ex3 = np.array([1, 1])
```

```python
result_ex1 = np.dot(iInv, ex1)
result_ex2 = np.dot(iInv, ex2)
result_ex3 = np.dot(iInv, ex3)
```

```python
print (result_ex1)
print (result_ex2)
print (result_ex3)
```

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits

import seaborn as sns
sns.set()
sns.set_style("whitegrid")
sns.set_color_codes()
%matplotlib inline
```

```python
f = sp.misc.face(gray=True)
plt.imshow(f, cmap = mpl.cm.bone)
plt.grid(False)
plt.show()
```

```python
import scipy.ndimage
```

```python
A = np.array([[1, 0.5], [0.3, 1]])
A
```

```python
Ainv = np.linalg.inv(A)
```

```python
f1 = sp.ndimage.affine_transform(f, Ainv)
plt.imshow(f1, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```
