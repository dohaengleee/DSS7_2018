```python
import requests, pymongo
```

```python
# client 연결
client = pymongo.MongoClient("mongodb://localhost:27017/")
client
```

```python
# database 연결
db = client.dss
db
```

```python
# show collections
db.collection_names()
```

```python
# select collection
collection = db.info
collection
```

### Find

```python
# 한개의 document 가져오기
document = collection.find_one({"subject":"css"})
document
```

```python
# 여러개의 document 가져오기
documents = collection.find()
document
for document in documents:
    print(document)
```

```python
# document의 개수를 확인
documents.count()
```

```python
# level이 3이하인 document를 내림차순으로 정렬
documents = collection.find({"level":{"$lte":3}}).sort("level", pymongo.DESCENDING) # 이상은 gte
for document in documents:
    print(document)
```

### Insert

```python
# 1개의 document 추가
data = {"subject":"css", "level":1, "comments":[{"name":"peter", "msg":"easy"}]}
result = collection.insert_one(data)
print(result.inserted_id)
```

```python
# 여러개의 document 추가
datas = [
    {"subject":"java", "level":1, "comments":[{"name":"peter", "msg":"easy"}]},
    {"subject":"html", "level":2, "comments":[{"name":"syzu", "msg":"medium"}]},
    {"subject":"gulp", "level":3, "comments":[{"name":"IU", "msg":"hard"}]},
]

result = collection.insert_many(datas)
print(result.inserted_ids)
```
